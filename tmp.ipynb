```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("sentimentdataset_clean.xlsx")
```

```python
df
```

```python
mapping = dict()
```

```python
for _, row in df.iterrows():
    mapping[row['General_Sentiment']] = []

for _, row in df.iterrows():
    if row['Specific_Sentiment'] not in mapping[row['General_Sentiment']]:
        mapping[row['General_Sentiment']].append(row['Specific_Sentiment'])
```

```python
for key, value in mapping.items():
    formatted_list = ', '.join(f'"{item.lower()}"' for item in value)

    print(f'    elif sentiment in [{formatted_list}]:')
    print(f'        return "{key}"')
```
